```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = '/content/drive/MyDrive/Multilayer Perceptron/Data/Discharge_tierras.csv' # Update to your local path
db = pd.read_csv(path)
discharge = db.to_numpy()
```

```python
import numpy as np
import matplotlib.pyplot as plt

discharge = np.array(discharge)

# Define custom bins for <1, 1–10, 10–100, 100–1000, >1000
bins = np.array([0, 1, 10, 100, 1000, np.inf])
labels = ["<1", "1–10", "10–100", "100–1000", ">1000"]

# Compute histogram
counts, _ = np.histogram(discharge, bins=bins)

colors = ["#e74c3c", "#e67e22", "#f1c40f", "#27ae60", "#2980b9"]
plt.figure(figsize=(10, 6.5))
bars = plt.bar(labels, counts, color=colors, width=0.6, edgecolor="black")

# Labels above each bar
for bar, count in zip(bars, counts):
    if count > 0:
        plt.text(bar.get_x() + bar.get_width()/2, count + 0.5, str(count),
                 ha='center', va='bottom', fontsize=12)

plt.xlabel("Discharge (m$^3$/s)", fontsize=16)
plt.ylabel("Number of experimental breakthrough curves", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
fig_path = "/content/drive/MyDrive"
plt.savefig(f"{fig_path}/Fig2.pdf", bbox_inches="tight")
plt.show()
```
